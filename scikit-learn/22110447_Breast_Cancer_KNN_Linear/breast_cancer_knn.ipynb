```python
import pandas as pd
```

```python
# Read data
df = pd.read_csv("data.csv")
```

```python
df
```

```python
# Drop unnecessary columns
col = ["id", "Unnamed: 32"]
df = df.drop(col, axis=1)
```

```python
df.describe()
```

```python
df.head(5)
```

```python
# map diagnosis column from str to float
def binary_map(x):
    return x.map({"M": 1, "B": 0})


df[["diagnosis"]] = df[["diagnosis"]].apply(binary_map)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Split variables and dest
X = df.drop("diagnosis", axis=1)
y = df["diagnosis"]
```

```python
# Split train and test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Standard unit of all column
# z = (x - u) / s
# z: new value
# x: current value
# u: mean value
# s:standard deviation
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# Create and train model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
)
```

```python
# Calculate error
print("accuracy_score: ", accuracy_score(y_test, y_pred) * 100)
print("precision_score: ", precision_score(y_test, y_pred) * 100)


print("recall_score: ", recall_score(y_test, y_pred) * 100)
print("f1_score: ", f1_score(y_test, y_pred) * 100)
```
