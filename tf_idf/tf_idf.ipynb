<h1>Preprocessing data</h1>


```python
data = [x.lower() for x in data]
data = [x.replace(",", "").replace(".", "").replace("!", "").replace("?", "") for x in data]

def remove_stopwords(sentence, stopwords):
    words = sentence.split()
    cleaned_words = [word for word in words if word.lower() not in stopwords]
    return ' '.join(cleaned_words)

def read_text_file(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()
            return content
    except FileNotFoundError:
        print("Không tìm thấy file:", filename)
        return None

filename = "vietnamese_stopword.txt"
vietnamese_stopwords = read_text_file(filename)
# https://www.kaggle.com/datasets/heeraldedhia/stop-words-in-28-languages?select=vietnamese.txt
data = [remove_stopwords(sentence, vietnamese_stopwords) for sentence in data]
data
```

<h1>Create split data</h1>


```python
split_data = []
for line in data:
    split_data.append(line.split())
print(split_data)
```

<h1>Find common word in all sentences. </1>


```python
common_word = set()
for line in split_data:
    for word in line:
        common_word.add(word)
common_word
```

<h1>Count word frequency of each sentence in split data by common word.</h1>


```python
import pandas as pd
word_frequency_data = []
for line in split_data:
    word_dict = dict.fromkeys(common_word, 0)
    for word in line:
        word_dict[word] += 1
    word_frequency_data.append(word_dict)
print(pd.DataFrame(word_frequency_data).T) 
```

<h1>Compute tf</h1>


```python
def compute_tf(word_dict, split_data):
    tf_dict = {}
    split_count = len(split_data)
    for word, count in word_dict.items():
        tf_dict[word] = count / float(split_count)
    return tf_dict
    
```

```python
for i in range(len(word_frequency_data)):
    print(compute_tf(word_frequency_data[i], split_data[i]))
```

<h1>Compute idf</h1>


```python
def compute_idf(data):
    import math
    idf_dict = {}
    N = len(data)
    
    idf_dict = dict.fromkeys(data[0].keys(), 0)
    for line in data:
        for word, val in line.items():
            if val > 0:
                idf_dict[word] += 1
                
    for word, val in idf_dict.items():
        idf_dict[word] = math.log10(N / float(val))
    return idf_dict
```

```python
print(compute_idf(word_frequency_data))
```

<h1>Compute tf idf</h1>


```python
def compute_tf_idf(tf, idfs):
    tf_idf = {}
    for word, val in tf.items():
        tf_idf[word] = val * idfs[word]
    return tf_idf
```

```python
res = []
for i in range(len(word_frequency_data)):
    tf = compute_tf(word_frequency_data[i], split_data[i])
    idfs = compute_idf(word_frequency_data)
    print(compute_tf_idf(tf, idfs))
    res.append(compute_tf_idf(tf, idfs))
pd.DataFrame(res).to_csv('tf_idf.csv', sep=',', encoding='utf-8-sig', header=True, index=False)
```

```python
def max_tf_idf(tf_idf):
    max_tf_idf = max(tf_idf.values())
    for word, val in tf_idf.items():
        if val == max_tf_idf:
            return word, val
for i in range(len(word_frequency_data)):
    print(max_tf_idf(res[i]))
```
