```python
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.1)
y = np.sin(x)
plt.plot(x, y, 'ro--', label='y=x^2', linewidth=2)
plt.legend(loc='upper right')
plt.xlabel("x", loc='right')
plt.ylabel("y", loc='top', rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x, y, width=0.05, color='blue', edgecolor='black')
plt.show()
```

```python
plt.pie([1, 2, 3, 4], labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%')
plt.show()
```

```python
plt.scatter(x, np.sin(x), c=np.log10(x), marker='o', cmap='viridis')
plt.colorbar(label='log$_{10}(x)$')
plt.show()  
```

```python
plt.violinplot([np.random.normal(0, std, 100) for std in range(1, 5)], showmeans=True)
plt.show()
```

```python
plt.polar(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu mẫu
labels = ['Nhóm 1', 'Nhóm 2', 'Nhóm 3']
men_means = [20, 35, 30]
women_means = [25, 32, 34]

x = np.arange(len(labels))  # vị trí các nhãn trên trục x
width = 0.35  # chiều rộng của các cột

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Nam')
rects2 = ax.bar(x + width/2, women_means, width, label='Nữ')

# Thêm các chi tiết
ax.set_xlabel('Nhóm')
ax.set_ylabel('Điểm số')
ax.set_title('Điểm số theo nhóm và giới tính')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Dữ liệu mẫu
labels = ['Nhóm 1', 'Nhóm 2', 'Nhóm 3']
men_means = [20, 35, 30]
women_means = [25, 32, 34]

x = np.arange(len(labels))  # vị trí các nhãn trên trục x

fig, ax = plt.subplots()
ax.bar(x, men_means, width=0.35, label='Nam')
ax.bar(x, women_means, width=0.35, bottom=men_means, label='Nữ')

# Thêm các chi tiết
ax.set_xlabel('Nhóm')
ax.set_ylabel('Điểm số')
ax.set_title('Điểm số theo nhóm và giới tính')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
