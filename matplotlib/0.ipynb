```python
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.1)
y = np.sin(x)
plt.plot(x, y, 'ro--', label='y=x^2', linewidth=2)
plt.legend(loc='upper right')
plt.xlabel("x", loc='right')
plt.ylabel("y", loc='top', rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x, y, width=0.05, color='blue', edgecolor='black')
plt.show()
```

```python
plt.pie([1, 2, 3, 4], labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%')
plt.show()
```

```python
plt.scatter(x, np.sin(x), c=np.log10(x), marker='o', cmap='viridis')
plt.colorbar(label='log$_{10}(x)$')
plt.show()  
```

```python
plt.violinplot([np.random.normal(0, std, 100) for std in range(1, 5)], showmeans=True)
plt.show()
```

```python
plt.polar(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# D liu mu
labels = ['Nhm 1', 'Nhm 2', 'Nhm 3']
men_means = [20, 35, 30]
women_means = [25, 32, 34]

x = np.arange(len(labels))  # v tr cc nhn trn trc x
width = 0.35  # chiu rng ca cc ct

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Nam')
rects2 = ax.bar(x + width/2, women_means, width, label='N')

# Thm cc chi tit
ax.set_xlabel('Nhm')
ax.set_ylabel('im s')
ax.set_title('im s theo nhm v gii tnh')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# D liu mu
labels = ['Nhm 1', 'Nhm 2', 'Nhm 3']
men_means = [20, 35, 30]
women_means = [25, 32, 34]

x = np.arange(len(labels))  # v tr cc nhn trn trc x

fig, ax = plt.subplots()
ax.bar(x, men_means, width=0.35, label='Nam')
ax.bar(x, women_means, width=0.35, bottom=men_means, label='N')

# Thm cc chi tit
ax.set_xlabel('Nhm')
ax.set_ylabel('im s')
ax.set_title('im s theo nhm v gii tnh')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
