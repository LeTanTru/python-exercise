```python
gdp_thuc_te_theo_quy = gdp_thuc_te_theo_quy.drop(columns=['Unnamed: 0'], axis=1)
gdp_thuc_te_theo_quy
```

```python
columns = gdp_thuc_te_theo_quy.columns
columns
```

```python
gdp_thuc_te = gdp_thuc_te_theo_quy.iloc[0:1, :].values.tolist()[0]
gdp_thuc_te
```

```python
_gdp_thuc_te = [int(x.replace(',', '')) for x in gdp_thuc_te]
plt.figure(figsize=(20, 10))
bars = plt.bar(list(columns), list(_gdp_thuc_te), color='b')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f"{yval:,}", ha='center', va='bottom', fontsize=8)

plt.gca().yaxis.set_major_locator(plt.MultipleLocator(100000))

def y_formatter(x, pos):
    return f'{int(x)}' 

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(y_formatter))

plt.xlabel('QuÃ½', fontsize=14)
plt.ylabel('Tá»•ng sáº£n pháº©m trong nÆ°á»›c theo giÃ¡ so sÃ¡nh', fontsize=14)
plt.title('Biá»ƒu Ä‘á»“ tá»•ng sáº£n pháº©m trong nÆ°á»›c theo giÃ¡ so sÃ¡nh theo quÃ½', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
_gdp_thuc_te = [int(x.replace(',', '')) for x in gdp_thuc_te]
fig, ax = plt.subplots(figsize=(20, 10))
bars = ax.bar(columns, _gdp_thuc_te, color='#0066ff')

for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 15000, f"{yval:,}", ha='center', va='bottom', fontsize=10)

ax.yaxis.set_major_locator(plt.MultipleLocator(100000))

def y_formatter(x, pos):
    return f'{int(x):,}'

ax.yaxis.set_major_formatter(ticker.FuncFormatter(y_formatter))

ax.annotate('Tá»· Ä‘á»“ng', xy=(0, 1), xytext=(0, 10),
            xycoords='axes fraction', textcoords='offset points',
            ha='center', va='bottom', fontsize=12, color='black')

ax.annotate('QuÃ½', xy=(1, 0), xytext=(0, -20),
            xycoords='axes fraction', textcoords='offset points',
            ha='center', va='bottom', fontsize=12, color='black')

plt.xticks(rotation=45, ha='center')
plt.tight_layout()
plt.grid(True, axis='y', linestyle='--', linewidth=0.5, color='black')
circle_marker = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#0066ff', markersize=10, label='GDP thá»±c táº¿')
ax.legend(handles=[circle_marker], loc='upper left')
plt.show()
```
