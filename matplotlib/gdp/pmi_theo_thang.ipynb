```python
pmi_theo_thang = pmi_theo_thang.drop(columns=['Unnamed: 0'], axis=1)
pmi_theo_thang
```

```python
columns = pmi_theo_thang.columns
columns
```

```python
_pmi_theo_thang = pmi_theo_thang.iloc[0:1, :].values.tolist()[0]
_pmi_theo_thang
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(columns, _pmi_theo_thang, marker='o', color='#16058b', linestyle='-')

for i, value in enumerate(_pmi_theo_thang):
    offset = 0.2 if value > 0 else -0.5
    ax.text(i, value + offset, f"{value:,}", ha='center', va='bottom', fontsize=10)

ax.yaxis.set_major_locator(plt.MultipleLocator(1))

def y_formatter(x, pos):
    return f'{int(x):,}'

ax.yaxis.set_major_formatter(ticker.FuncFormatter(y_formatter))

ax.annotate('PMI', xy=(0, 1), xytext=(0, 5),
            xycoords='axes fraction', textcoords='offset points',
            ha='center', va='bottom', fontsize=20, color='#0066ff')

ax.annotate('QuÃ½', xy=(1, 0), xytext=(0, -20),
            xycoords='axes fraction', textcoords='offset points',
            ha='center', va='bottom', fontsize=12, color='black')

plt.xticks(ha='center', rotation=45)
plt.tight_layout()
plt.grid(True, axis='y', linestyle='--', linewidth=0.5, color='gray')
plt.legend(['PMI'], loc='upper right', fontsize=12)
plt.show()
```
