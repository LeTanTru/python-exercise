```python
toc_do_tang_truong_gdp_thuc_te = toc_do_tang_truong_gdp_thuc_te.drop(columns=['Unnamed: 0'], axis=1)
toc_do_tang_truong_gdp_thuc_te
```

```python
columns = toc_do_tang_truong_gdp_thuc_te.columns
columns
```

```python
gdp_thuc_te = toc_do_tang_truong_gdp_thuc_te.iloc[0:1, :].values.tolist()[0]
gdp_thuc_te
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(columns, gdp_thuc_te, marker='o', color='#0066ff', linestyle='-')

for i, value in enumerate(gdp_thuc_te):
    offset = 0.2 if value > 0 else -0.5
    ax.text(i, value + offset, f"{value:,}", ha='center', va='bottom', fontsize=8)

ax.yaxis.set_major_locator(plt.MultipleLocator(1))

def y_formatter(x, pos):
    return f'{int(x):,}'

ax.yaxis.set_major_formatter(ticker.FuncFormatter(y_formatter))

ax.set_xlabel('Qu', fontsize=14)
ax.set_ylabel('Tng sn phm trong nc theo gi so snh', fontsize=14)

ax.annotate('%', xy=(0, 1), xytext=(0, 5),
            xycoords='axes fraction', textcoords='offset points',
            ha='center', va='bottom', fontsize=12, color='black')

ax.annotate('Qu', xy=(1, 0), xytext=(0, -20),
            xycoords='axes fraction', textcoords='offset points',
            ha='center', va='bottom', fontsize=12, color='black')

plt.xticks(ha='center')
plt.tight_layout()
plt.grid(True, axis='y', linestyle='--', linewidth=0.5, color='gray')
circle_marker = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#0066ff', markersize=10, label='Tc  tng trng')
ax.legend(handles=[circle_marker], loc='upper left')
plt.show()
```
